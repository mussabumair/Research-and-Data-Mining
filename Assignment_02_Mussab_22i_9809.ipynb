```python
df['PoolQC'].fillna(df['PoolQC'].mode()[0], inplace = True)
df['Alley'].fillna(df['Alley'].mode()[0], inplace = True)
df['Fence'].fillna(df['Fence'].mode()[0], inplace = True)
df['MiscFeature'].fillna(df['MiscFeature'].mode()[0], inplace= True)
```

```python
df['LotFrontage'].fillna(df['LotFrontage'].mode()[0], inplace= True)
```

```python
df.drop(['Id'], axis=1,inplace=True)
df.head()
```

```python
df['Overall Rating'] = (df['OverallQual'] + df['OverallCond']) /2
df.head()
#combined the two columns of overall quality and overall rating and then divided them by 2 to give an overall rating of quality and condition in one column
```

```python
df['Normalized_Sale_Price'] = (df['SalePrice'] - df['SalePrice'].min()) / (df['SalePrice'].max() - df['SalePrice'].min())
```

```python
df['LotArea'] = pd.cut(df['LotArea'], bins=[0, 2000, 5000, 6000, 8000], labels=['Small', 'Average', 'Medium', 'Large'])
```

```python
print(df.sort_values('SalePrice', ascending=False).head())
```

```python
fireplaces_counter = df[(df['MSSubClass'] == 60) & (df['SaleCondition'] == 'Normal')]
print("Number of fireplaces in 60 MS Sub Class:", len(fireplaces_counter))
```

```python
def sale_category(SalePrice):
  if SalePrice < 100000:
      return 'Low'
  elif SalePrice < 180000:
      return 'Medium'
  else:
      return 'High'

df['sale_category'] = df['SalePrice'].apply(sale_category)
```

```python
print(df['sale_category'])
```

```python
print("\n23. Reshaping Data")
melted_df = pd.melt(df, id_vars=['MSSubClass', 'Street'], value_vars=['SalePrice', 'Overall Rating'])
print(melted_df.head())
```
